# Gender Classification Model
In this notebook, I added a gender prediction to the tweets dataset. I first cleaned the data and transformed it into a bag-of-words model using sklearn methods. Then, I tested different models such as random forest, logistic regression models, and a Multinomial Naive Bayes model. This was done in addition to the SVM trained in the previous notebook, but the proccess was streamlined by using sklearn methods for ease of understanding and application.

In order to train my own model, I first had to source twitter data which was gender labeled (as in, it was written by a person of a certain gender). I found relevant data on a Kaggle competition on an open source dataset, which can be found here:https://www.kaggle.com/crowdflower/twitter-user-gender-classification
Though there is also image, username and other data included in this set, I only used the gender label and the sample tweet. The confidence level of the gender label was also used to sort out which data points would be useful for training our model. I decided that tweets with confidence level greater than .8 were to be used for training. This was because setting confidence threshold to 1 proved to return too few data points for training.

```python
labeled_dat = pd.read_csv("gender_classified.csv", sep=",", engine='python')

# I only want the single text sample and the gender label for my analysis.
# The data set also included a confidence level of the gender label, of which I only chose the ones with greater than 80%
training_data = labeled_dat[labeled_dat['gender:confidence'] > .8][['gender', 'text', 'gender:confidence']]
training_data.head()

# This results in 13926 samples where the assigned gender category of the writer is certain.
training_data.shape
training_data['gender'].value_counts()
```

```python
# Getting rid of the unknown labeled data
training_data = training_data[training_data['gender'] != 'unknown']
training_data['text'] = training_data['text'].str.lower() # Lower Case
training_data['text'].isnull().values.any()
```

This dataset included not just male and female tweets, but also tweets by brand twitter accounts and some which weere unknown. I removed tweets for which the gender was unknown, and decided to use two dummy variables to encode the gender. One dummy variable was 1 for male and 0 for non-male (i.e. female or brand). A similar dummy variable was used for females. 

Originally, I had tried a model with 1 as male and 0 as female, but this produced very low accuracy rates. This was due to the fact that removing the brand and unknown data greatly reduced the dataset size, but also because using two separate models provided greater accuracy.

```python
training_encode_m = training_data.copy()[['text', 'gender']]
training_encode_m['gender'] = [1 if i == 'male' else 0 for i in training_data['gender'].values]
```

```python
training_encode_f = training_data.copy()[['text', 'gender']]
training_encode_f['gender'] = [1 if i == 'female' else 0 for i in training_data['gender'].values]
```

```python
vectorizer = feature_extraction.text.CountVectorizer(strip_accents='ascii', ngram_range=(1, 2))
X_all = vectorizer.fit_transform(training_data['text'])
```

Training a random forest.

```python
random_forest_m = ensemble.RandomForestClassifier(n_estimators=200,max_depth=50)
```

```python
# Splitting train and test data for males
y_m = training_encode_m['gender'].values
mX_train, mX_test, my_train, my_test = sklearn.model_selection.train_test_split(X_all, y_m, test_size=0.20, random_state=84)
```

```python
# Splitting train and test data for females
y_f = training_encode_f['gender'].values
fX_train, fX_test, fy_train, fy_test = sklearn.model_selection.train_test_split(X_all, y_m, test_size=0.20, random_state=59)
```

```python
random_forest_m.fit(mX_train, my_train)
rf_m = random_forest_m.predict(mX_test)
metrics.accuracy_score(my_test, rf_m)
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(my_test, rf_m).ravel()
(tn, fp, fn, tp)
```

```python
plt.matshow(metrics.confusion_matrix(my_test, rf_m))
plt.colorbar()
```

For the male model, it is easier to find true negatives than true positives. In fact, it seems like the majority of points are getting classified as negative, and the actual number of true positives is quite small. 

```python
random_forest_f = ensemble.RandomForestClassifier(n_estimators=200,max_depth=50)
```

```python
random_forest_f.fit(fX_train, fy_train)
rf_f = random_forest_f.predict(fX_test)
metrics.accuracy_score(fy_test, rf_f)
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(fy_test, rf_f).ravel()
(tn, fp, fn, tp)
```

```python
plt.matshow(metrics.confusion_matrix(fy_test, rf_f))
plt.colorbar()
```

In general, the random forest models seem to be good at telling what isn't the right class, but pretty bad at telling what is the right class. So for example, for the male random forest, it seems like the false negatives are very high, but it categorizes 936 males inappropriately in the non-male class. A similar result can be observed for the females.

```python
# importing all necessery modules 
from wordcloud import WordCloud, STOPWORDS 
```

```python
# Code from https://www.geeksforgeeks.org/generating-word-cloud-python/
# Encoded into function
def word_cloud_gen(dataframe):
    comment_words = '' 
    stopwords = set(STOPWORDS) 
    url_re = r'http'
    co_re = r'co'
    rt_re = r'rt'
    
    # iterate through the csv file 
    for text in dataframe['text']: 
          
        # typecaste each val to string 
        text = str(text)
        text = text.replace(url_re, '') # Remove Links/URL
        text = text.replace(co_re, '')
        text = text.replace(rt_re, '')
        
        # split the value 
        tokens = text.split() 
          
        # Converts each token into lowercase 
        for i in range(len(tokens)): 
            tokens[i] = tokens[i].lower() 
          
        comment_words += " ".join(tokens)+" "
      
    wordcloud = WordCloud(width = 800, height = 800, 
                    background_color ='white', 
                    stopwords = stopwords, 
                    min_font_size = 10).generate(comment_words) 
      
    # plot the WordCloud image                        
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
      
    plt.show() 
```
